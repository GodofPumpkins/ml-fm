```python
# spiegelib synth configuration with all 6 operators, all envelope parameters allowed

import dawdreamer as daw
from scipy.io import wavfile
import numpy as np

import tensorflow as tf
import keras.api._v2.keras as keras
import keras.layers as layers

import os.path

#import local edited version of spiegelib

from pathlib import Path
import sys

path_root = os.path.abspath('')
sys.path.append(str(path_root))
print(sys.path)

import spiegelib as spgl


#declare variables
SAMPLE_RATE = 44100
BUFFER_SIZE = 128
DURATION = 3 # overall duration of each audio file in seconds. This limits the range of attack values that will have an effect on the 
NOTE_DURATION = 2 # duration of midi note on message
START = 0

SYNTH_PLUGIN = "C:/FYP/vst2 dexed/Dexed.dll"

targets = spgl.AudioBuffer.load_folder('./evaluation_gen_final/audio')

synth = spgl.synth.SynthVST(SYNTH_PLUGIN,
                            note_length_secs=NOTE_DURATION,
                            render_length_secs=DURATION,
                            sample_rate=SAMPLE_RATE)

for i in range(len(targets)):
    synth.load_state("./synth_params/gen_patches/dexed_sample_%s.json" % i)
    synth.synth.open_editor()
```

```python
# create spectrogram comparisons of input and output audio files

import librosa as lb
import numpy as np
import matplotlib.pyplot as plt

filenames = ['alto_sax', 'bassoon', 'dx7_ep', 'hammond', 'lap_steel', 'marimba', 'minimoog_lead', 'ms20', 'solina', 'synth_sawtooth', 'trumpet', 'wurli']

for i in range(12):
    y_o, sr = lb.load('./audio_samples/audio/' + filenames[i] + '.wav')
    S_o = lb.feature.melspectrogram(y=y_o, sr=sr)
    y_pred, sr = lb.load('./audio_samples/nsga_final/nsga_prediction_' + str(i) + '.wav')
    S_pred = lb.feature.melspectrogram(y=y_pred, sr=sr)

    fig, (ax1, ax2) = plt.subplots(1, 2, width_ratios=[1, 1.2])
    fig.set_figwidth(13)
    fig.set_figheight(4)
    S_dB_o = lb.power_to_db(S_o, ref=np.max)
    img1 = lb.display.specshow(S_dB_o, x_axis='time',
                            y_axis='mel', sr=sr,
                            fmax=sr, ax=ax1)
    ax1.set(title='Sample: ' + filenames[i])

    S_dB_pred = lb.power_to_db(S_pred, ref=np.max)
    img2 = lb.display.specshow(S_dB_pred, x_axis='time',
                            y_axis='mel', sr=sr,
                            fmax=sr, ax=ax2)
    fig.colorbar(img2, ax=ax2, format='%+2.0f dB')
    ax2.set(title='Predicted Output')
    
    fig.show()
```
